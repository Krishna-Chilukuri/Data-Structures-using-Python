```python
class GraphsAM:
    def __init__(self, nodes) -> None:
        self.__indexNodes = {}
        i = 0
        for node in nodes:
            self.__indexNodes[node] = i
            i += 1
        self.__adjMat = [[0 for i in range(len(nodes))] for i in range(len(nodes))]
        for i in range(len(self.__adjMat)):
            print(self.__adjMat[i])
        # print(self.__adjMat, sep = "\n")

    def addEdges(self, edges):
        for edge in edges:
            self.__adjMat[self.__indexNodes[edge[0]]][self.__indexNodes[edge[1]]] = 1
    
    def addEdge(self, edge):
        self.__adjMat[self.__indexNodes[edge[0]]][self.__indexNodes[edge[1]]] = 1

    def removeEdge(self, edge):
        if self.__adjMat[self.__indexNodes[edge[0]]][self.__indexNodes[edge[1]]] == 1:
            self.__adjMat[self.__indexNodes[edge[0]]][self.__indexNodes[edge[1]]] = 0
    
    #Traversals - BFS & DFS
    def bfs(self, startingVertex):
        nextQ = []
        visited = set()
        nextQ.append(startingVertex)

        while nextQ:
            currentVertex = nextQ.pop(0)
            if currentVertex not in visited:
                visited.add(currentVertex)
                print(currentVertex, end = " ")
                i = 0
                for neighbor in self.__adjMat[self.__indexNodes[currentVertex]]:
                    if neighbor == 1:
                        key = list(filter(lambda x: self.__indexNodes[x] == i, self.__indexNodes))[0]
                        nextQ.append(key)
                    i += 1

    def dfs(self, startingVertex):
        nextStack = []
        visited = set()
        nextStack.append(startingVertex)

        while nextStack:
            currentVertex = nextStack.pop()
            if currentVertex not in visited:
                visited.add(currentVertex)
                print(currentVertex, end = " ")
                i = 0
                for neighbor in self.__adjMat[self.__indexNodes[currentVertex]]:
                    if neighbor == 1:
                        key = list(filter(lambda x: self.__indexNodes[x] == i, self.__indexNodes))[0]
                        nextStack.append(key)
                    i += 1

    def printer(self):
        print(f"Vertices: {self.__indexNodes}")
        for i in range(len(self.__adjMat)):
            print(self.__adjMat[i])



if __name__ == "__main__":
    graph1 = GraphsAM(['A', 'B', 'C', 'D', 'E'])
    edges = [('A', 'B'), ('B', 'A'), ('C', 'A'), ('C', 'D'), ('D', 'E')]
    graph1.addEdges(edges)
    graph1.printer()
    graph1.addEdge(('B', 'C'))
    graph1.printer()
    graph1.removeEdge(('A', 'B'))
    graph1.printer()
    print("DFS Traversal: ", end = "")
    graph1.dfs('B')
    print("\nBFS Traversal: ", end = "")
    graph1.bfs('B')
```
