```python
#Binary Tree implementation
from collections import deque
class binaryTree:

    def __init__(self, val) -> None:
        self.val = val
        self.left = None
        self.right = None
    
    def insertNode(self, val, insertQ = []):
        stat = False
        # In such applications, recursion is an overhead. Looping is better in this case
        if self.left == None:
            self.left = binaryTree(val)
            return
        else:
            insertQ.append(self.left)
        if self.right == None:
            self.right = binaryTree(val)
            return
        else:
            insertQ.append(self.right)
        
        popp = insertQ.pop(0)
        popp.insertNode(val, insertQ)
        

    def deleteNode(self, val):
        if self == None:
            return None
        if self.left == None and self.right == None:
            if self.val == val:
                return None
            else:
                return self
        del_node = None
        delQ = deque()
        delQ.append((self, None))
        stat = False
        while delQ:
            node, par = delQ.popleft()
            if node.val == val:
                del_node = node
                stat = True
            if node.left:
                delQ.append((node.left, node))
            if node.right:
                delQ.append((node.right, node))
        
        if not stat:
            print("Node not found")
        else:
            temp = node.val
            if par.left and par.left.val == temp:
                par.left = None
            elif par.right and par.right.val == temp:
                par.right = None
            del_node.val = temp
        return self
    

def inorderTraversal(node):
        if not node:
            return
        inorderTraversal(node.left)
        print(node.val, end = " ")
        inorderTraversal(node.right)

def preorderTraversal(node):
    if not node:
        return
    print(node.val, end=" ")
    preorderTraversal(node.left)
    preorderTraversal(node.right)

def postorderTraversal(node):
    if not node:
        return
    postorderTraversal(node.left)
    postorderTraversal(node.right)
    print(node.val, end = " ")
    

def mySwitch(rootNode, x):
        if x == 1:
            val = int(input("Enter value to insert: "))
            rootNode.insertNode(val)
        elif x == 2:
            val = int(input("Enter value to delete: "))
            rootNode = rootNode.deleteNode(val)
        elif x == 3:
            inorderTraversal(rootNode)
        elif x == 4:
            preorderTraversal(rootNode)
        elif x == 5:
            postorderTraversal(rootNode)
        else:
            print("Select a correct choice")

rootNode = binaryTree(int(input("Enter root Value: ")))
if __name__ == "__main__":
    print(rootNode.val)
    while True:
        print()
        print("1. Insert")
        print("2. Delete")
        print("3. InOrder traversal")
        print("4. PreOrder traversal")
        print("5. PostOrder traversal")
        print("6. Exit")
        x1 = int(input("Enter your Choice: "))
        if x1 == 6:
            break
        mySwitch(rootNode, x1)

```
