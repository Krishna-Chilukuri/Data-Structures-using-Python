```python
class GraphsAL:
    def __init__(self, nodes) -> None:
        self.__nodes = nodes
        self.__adjList = {}
        for node in self.__nodes:
            self.__adjList[node] = []

    def addEdges(self, edges):
        for edge in edges:
            self.__adjList[edge[0]].append(edge[1])
    
    def addEdge(self, edge):
        self.__adjList[edge[0]].append(edge[1])

    def removeEdge(self, edge):
        if edge[1] in self.__adjList[edge[0]]:
            self.__adjList[edge[0]].remove(edge[1])
    
    #Traversals - BFS & DFS
    # def recursiveDFS(self, vertex, visited):
    #     visited.add(vertex)
    #     print(vertex, end = " ")
    #     # print(f"Neighbors of {vertex} are {self.__adjList[vertex]}")
    #     for neighbor in self.__adjList[vertex]:
    #         if neighbor not in visited:
    #             self.recursiveDFS(neighbor, visited)

    # def dfs(self, startingVertex):
    #     visited = set()
    #     self.recursiveDFS(startingVertex, visited)
    def bfs(self, startingVertex):
        nextQ = []
        visited = set()
        nextQ.append(startingVertex)

        while nextQ:
            currentVertex = nextQ.pop(0)
            if currentVertex not in visited:
                visited.add(currentVertex)
                print(currentVertex, end = " ")
                for neighbor in self.__adjList[currentVertex]:
                    nextQ.append(neighbor)

    def dfs(self, startingVertex):
        nextStack = []
        visited = set()
        nextStack.append(startingVertex)

        while nextStack:
            currentVertex = nextStack.pop()
            if currentVertex not in visited:
                visited.add(currentVertex)
                print(currentVertex, end = " ")
                for neighbor in self.__adjList[currentVertex]:
                    nextStack.append(neighbor)

    def printer(self):
        print(f"Vertices: {self.__nodes}\nAdjacency List: {self.__adjList}")



if __name__ == "__main__":
    graph1 = GraphsAL(['A', 'B', 'C', 'D', 'E'])
    edges = [('A', 'B'), ('B', 'A'), ('C', 'A'), ('C', 'D'), ('D', 'E')]
    graph1.addEdges(edges)
    graph1.printer()
    graph1.addEdge(('B', 'C'))
    graph1.printer()
    graph1.removeEdge(('A', 'B'))
    graph1.printer()
    print("DFS Traversal: ", end = "")
    graph1.dfs('B')
    print("\nBFS Traversal: ", end = "")
    graph1.bfs('B')
```
