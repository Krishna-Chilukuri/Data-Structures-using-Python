```python
class Node:
    def __init__(self, val) -> None:
        self.val = val
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self, val) -> None:
        self.head = Node(val)
        # self.head = None
    def insertAtFront(self, val):
        newNode = Node(val)
        if self.head is None:
            self.head = newNode
        else:
            newNode.next = self.head
            self.head.prev = newNode
            self.head = newNode
    def insertAtEnd(self, val):
        newNode = Node(val)
        if self.head is None:
            self.head = newNode
        else:
            temp = self.head
            while(temp.next):
                temp = temp.next
            temp.next = newNode
            newNode.prev = temp
    def insertAtPos(self, val, index):
        locIndex = 1
        newNode = Node(val)
        temp = self.head
        while temp and locIndex != index:
            temp = temp.next
            locIndex += 1
        if temp != None:
            prev = temp.prev
            prev.next = newNode
            newNode.prev = prev
            newNode.next = temp
            temp.prev = newNode
        else:
            print(f"Insertion at {index} is not possible")
        

    def remove(self, val):
        if self.head.val == val:
            if self.head.next:
                headNext = self.head.next
                headNext.prev = None
                self.head.next = None
                self.head = headNext
            else:
                self.head = None
        else:
            temp = self.head.next
            while temp:
                if temp.val == val:
                    prev = temp.prev
                    nxt = temp.next
                    if nxt:
                        prev.next = nxt
                        nxt.prev = prev

                        temp.next = None
                        temp.prev = None
                    else:
                        prev.next = None
                        temp.prev = None
                    break
                temp = temp.next
            else:
                print(f"{val} not found to remove!")
    def search(self, val):
        temp = self.head
        while temp:
            if temp.val == val:
                print(f"{val} found!!!")
                break
            temp = temp.next
        else:
            print(f"{val} not found!!!")
    def replace(self, oldVal, updatedVal):
        temp = self.head
        while temp:
            if temp.val == oldVal:
                temp.val = updatedVal
                print("Value updated!")
                break
            temp = temp.next
        else:
            print(f"{oldVal} not found to replace!")

    def printer(self):
        temp = self.head
        while temp:
            print(temp.val, end=" <=> ")
            temp = temp.next
        print("END")

linkedList = DoublyLinkedList(10)
linkedList.insertAtEnd(20)
linkedList.insertAtFront(30)
linkedList.insertAtEnd(40)
linkedList.printer()
linkedList.remove(30)
linkedList.printer()
linkedList.remove(20)
linkedList.printer()
linkedList.search(40)
linkedList.remove(50)
linkedList.replace(40, 50)
linkedList.printer()
```
