```python
class Node:
    def __init__(self, val) -> None:
        self.val = val
        self.next = None

class LinkedList:
    def __init__(self, val) -> None:
        self.__head = Node(val)
        # self.head = None
    def insertAtFront(self, val):
        newNode = Node(val)
        if self.__head is None:
            self.__head = newNode
        else:
            newNode.next = self.__head
            self.__head = newNode
    def insertAtEnd(self, val):
        newNode = Node(val)
        if self.__head is None:
            self.__head = newNode
        else:
            temp = self.__head
            while(temp.next):
                temp = temp.next
            temp.next = newNode
    def insertAtPos(self, val, index):
        locIndex = 1
        newNode = Node(val)
        temp = self.__head
        while temp and locIndex != index:
            temp = temp.next
            locIndex += 1
        if temp != None:
            newNode.__ext = temp.next
            temp.next = newNode
        else:
            print(f"Insertion at {index} is not possible")
        

    def remove(self, val):
        if self.__head.val == val:
            self.__head = self.__head.next
        else:
            temp = self.__head.next
            prev = self.__head
            while temp:
                if temp.val == val:
                    prev.next = temp.next
                    temp.next = None
                    break
                temp = temp.next
                prev = prev.next
            else:
                print(f"{val} not found to remove!")
    def search(self, val):
        temp = self.__head
        while temp:
            if temp.val == val:
                print(f"{val} found!!!")
                break
            temp = temp.next
        else:
            print(f"{val} not found!!!")
    def replace(self, oldVal, updatedVal):
        temp = self.__head
        while temp:
            if temp.val == oldVal:
                temp.val = updatedVal
                print("Value updated!")
            temp = temp.next
        else:
            print(f"{oldVal} not found to replace!")

    def printer(self):
        temp = self.__head
        while temp:
            print(temp.val, end=" -> ")
            temp = temp.next
        print("END")

linkedList = LinkedList(10)
linkedList.insertAtEnd(20)
linkedList.insertAtFront(30)
linkedList.insertAtEnd(40)
linkedList.printer()
linkedList.remove(30)
linkedList.printer()
linkedList.remove(20)
linkedList.printer()
linkedList.search(40)
linkedList.remove(50)
linkedList.replace(40, 50)
linkedList.printer()
```
